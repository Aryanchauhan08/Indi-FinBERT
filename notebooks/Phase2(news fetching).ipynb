# DATA FETCHING (MONEYCONTROL)

```python
import requests
from bs4 import BeautifulSoup
import time

HEADERS = {"User-Agent": "Mozilla/5.0"}

BASE_URLS = [
    "https://www.moneycontrol.com/news/business/stocks/page-{}/",
    "https://www.moneycontrol.com/news/business/markets/page-{}/",
    "https://www.moneycontrol.com/news/business/economy/page-{}/",
    "https://www.moneycontrol.com/news/business/earnings/page-{}/",
    "https://www.moneycontrol.com/news/business/ipo/page-{}/",
]

JUNK_PHRASES = [
    "day’s trial", "subscribe", "sign up", "download the app",
    "advisory alert", "read also", "also read", "newsletter",
]

def is_real_headline(text):
    if len(text.split()) < 6:
        return False
    text_lower = text.lower()
    if any(phrase in text_lower for phrase in JUNK_PHRASES):
        return False
    return True

all_headlines = []
seen = set()

for base_url in BASE_URLS:
    for page in range(1, 31):
        url = base_url.format(page)
        response = requests.get(url, headers=HEADERS)
        soup = BeautifulSoup(response.content, 'html.parser')
        headlines = soup.find_all('h2')

        for headline in headlines:
            text = headline.text.strip()
            if not is_real_headline(text):
                continue
            if text in seen:
                continue
            seen.add(text)
            all_headlines.append(text)

        print(f"{url} → {len(all_headlines)} total so far")
        time.sleep(1)

print(f"\nTotal unique headlines: {len(all_headlines)}")
```
